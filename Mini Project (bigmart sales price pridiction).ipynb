```python
x = final_df.drop('Item_Outlet_Sales', axis = 1)
y = final_df['Item_Outlet_Sales']
```

# MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
nor_scaler = MinMaxScaler()
x_nor_df = nor_scaler.fit_transform(x)
x_df = pd.DataFrame(x_nor_df, columns = x.columns)
x_df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

# Train Test Split 

```python
x_train,x_test,y_train,y_test = train_test_split(x_df,y, test_size = 0.2, random_state = 1)

x_train.shape
```

# Train Model 

# linear regression

```python
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
y_pred = linear_model.predict(x_test)
y_pred
```

```python
y_test
```

# Evaluation

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test,y_pred)
```
