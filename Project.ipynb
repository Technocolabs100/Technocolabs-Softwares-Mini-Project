```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from  sklearn.preprocessing import OneHotEncoder

import numpy as np
```

```python
def preprocessing(df):
  # Deleting Extra Columns
  df1 = df.drop(df.loc[:,['Item_Identifier', 'Item_Fat_Content','Item_Weight','Outlet_Establishment_Year']], axis=1)

  # Imputing Values Catagorical
  cat = ['Item_Type', 'Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type']
  num = ['Item_Visibility','Item_MRP','Age','Item_Outlet_Sales']

  s_imputer = SimpleImputer(strategy='most_frequent',verbose=0)
  df1_cat = s_imputer.fit_transform(df1.loc[:,cat])
  df1_cat = pd.DataFrame(df1_cat)

  n_imputer = SimpleImputer(strategy='mean',verbose=0)
  df1_num = n_imputer.fit_transform(df1.loc[:,num])
  df1_num = pd.DataFrame(df1_num)

  df1_cat.rename(columns = {0 : 'Item_Type', 1 : 'Outlet_Identifier', 2 : 'Outlet_Size', 3 : 'Outlet_Location_Type', 4 :'Outlet_Type'} , inplace = True)
  df1_num.rename(columns = {0 : 'Item_Visibility', 1 : 'Item_MRP', 2 : 'Age',3 :'Item_Outlet_Sales'} , inplace = True)

  frames = [df1_cat, df1_num]
  result_imp = pd.concat(frames, axis = 1)

  
  ##################  ALL VALUES ARE IMPUTED UPTILL NOW #########################

  label_encoder = LabelEncoder()
  result_imp['Outlet_Size'] = label_encoder.fit_transform(result_imp['Outlet_Size'])

  ##################  ALL VALUES ARE LABEL ENCODED NOW #########################

  ohe = OneHotEncoder()

  feature_array1 = ohe.fit_transform(result_imp[['Outlet_Identifier']]).toarray()
  feature_labels1 = ohe.categories_
  d1 = pd.DataFrame(feature_array1, columns = feature_labels1)

  feature_array2 = ohe.fit_transform(result_imp[['Outlet_Location_Type']]).toarray()
  feature_labels2 = ohe.categories_
  d2 = pd.DataFrame(feature_array2, columns = feature_labels2)

  feature_array3 = ohe.fit_transform(result_imp[['Outlet_Type']]).toarray()
  feature_labels3 = ohe.categories_
  d3 = pd.DataFrame(feature_array3, columns = feature_labels3)

  feature_array4 = ohe.fit_transform(result_imp[['Item_Type']]).toarray()
  feature_labels4 = ohe.categories_
  d4 = pd.DataFrame(feature_array4, columns = feature_labels4)

  d1 = pd.DataFrame(d1)
  d2 = pd.DataFrame(d2)
  d3 = pd.DataFrame(d3)
  d4 = pd.DataFrame(d4)

  return d1, d2, d3, d4, result_imp

  
  
```

```python
d1, d2, d3, d4, result_imp = preprocessing(df)
```

```python
data = pd.concat([d1, d2, d3, d4, result_imp], axis=1, join='inner')
```

```python
newdf = data.drop(['Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type','Item_Type' ], axis='columns')
```
